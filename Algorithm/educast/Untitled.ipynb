```python
# 1 ~ 7  

can = [i for i in range(1,8)]
arraySize = 7
callsize = 0


def swap(i, j):
    can[i], can[j] = can[j], can[i]


def permu(arrSize):
    global callsize

    if arrSize == 1: # 
        callsize += 1
        print(*can)
        return

    for i in range(arrSize):
        swap(i, arrSize-1) # 
        permu(arrSize-1) # 
        swap(arrSize-1, i)

permu(arraySize)
print(callsize)
```

```python
#  

def spread_virus(walls):
    # 3   
    #           .
    #       

    global min_spreaded_virus

    chk_map = [[False] * m for _ in range(n)]
    for i, j in walls:
        chk_map[i][j] = True

    q = []
    spreaded_virus_cnt = 0
    for i, j in viruses:
        chk_map[i][j] = True
        q.append((i, j))
    # ,       
    #       .
    while q:
        sr, sc = q.pop()
        for d in range(4):
            nr = sr + dr[d]
            nc = sc + dc[d]

            if nr < 0 or nc < 0 or nr >= n or nc >= m: continue
            if chk_map[nr][nc] or a[nr][nc]: continue
            chk_map[nr][nc] = True
            spreaded_virus_cnt += 1
            q.append((nr, nc))

        if spreaded_virus_cnt > min_spreaded_virus:
            return

    min_spreaded_virus = min(min_spreaded_virus, spreaded_virus_cnt)

def build_three_walls(idx = 0, wall_cnt = 0, walls = []):
    if wall_cnt == 3:
        spread_virus(walls)
        return

    for zero in zeroes[idx:]:
        walls.append(zero)
        build_three_walls(idx+1, wall_cnt+1, walls)
        walls.pop()


n, m = map(int, input().split())

a = [list(map(int, input().split())) for _ in range(n)]

viruses = []
zeroes = []

for r in range(n):
    for c in range(m):
        if a[r][c] == 2:
            viruses.append((r, c))
        elif a[r][c] == 0:
            zeroes.append((r, c))

dr = [-1, 0, 1, 0]
dc = [0, 1, 0, -1]

min_spreaded_virus = 1e9

build_three_walls()

max_survived = len(zeroes) - 3 - min_spreaded_virus

print(max_survived)
```

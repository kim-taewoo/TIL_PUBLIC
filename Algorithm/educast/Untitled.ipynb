```python
# 1 ~ 7 까지 조합

can = [i for i in range(1,8)]
arraySize = 7
callsize = 0


def swap(i, j):
    can[i], can[j] = can[j], can[i]


def permu(arrSize):
    global callsize

    if arrSize == 1: # 종료부
        callsize += 1
        print(*can)
        return

    for i in range(arrSize):
        swap(i, arrSize-1) # 해결부
        permu(arrSize-1) # 분할부
        swap(arrSize-1, i)

permu(arraySize)
print(callsize)
```

```python
# 등산로 조성 swexpert 개선 전.
# k 가 '최대' 가능 깊이라는 것에 유의.

T = int(input())

dr = [-1, 0, 1, 0]
dc = [0, 1, 0, -1]


def dfs(sr, sc, length):
    global max_length
    global k_used
    max_length = max(max_length, length)
    for d in range(4):
        nr = sr + dr[d]
        nc = sc + dc[d]

        if nr < 0 or nc < 0 or nr >= n or nc >= n: continue
        if chk_map[nr][nc]: continue
        if a[nr][nc] >= a[sr][sc]:
            if k_used: continue
            else:
                for k_value in range(1, k+1):
                    if (a[nr][nc] - k_value) < a[sr][sc]:
                        k_used = True
                        a[nr][nc] = a[nr][nc] - k_value
                        chk_map[nr][nc] = True
                        dfs(nr, nc, length + 1)
                        a[nr][nc] = a[nr][nc] + k_value
                        chk_map[nr][nc] = False
                        k_used = False
        elif a[nr][nc] < a[sr][sc]:
            chk_map[nr][nc] = True
            dfs(nr, nc, length + 1)
            chk_map[nr][nc] = False


for t in range(T):
    n, k = map(int, input().split())
    a = [list(map(int, input().split())) for _ in range(n)]
    chk_map = [[False] * n for _ in range(n)]
    highest = 1
    coor_highest = []
    for r in range(n):
        for c in range(n):
            if a[r][c] > highest:
                coor_highest = [(r, c)]
                highest = a[r][c]
            elif a[r][c] == highest:
                coor_highest.append((r, c))

    max_length = 1
    k_used = False
    for r, c in coor_highest:
        chk_map[r][c] = True
        k_used = False
        dfs(r, c, 1)
        chk_map[r][c] = False

    print('#{} {}'.format(t+1, max_length))
```

```python
# 등산로 조금 깔끔하게 개선
# k 를 굳이 for 문 돌 필요가 없음을 깨달아야 한다. 
# 어쨌거나 커야지 더 좋은거니까 그냥 작아질 수 있다는 거 확인만 되면 -1 만큼만 작게 해준다.
# 그리고 bool 값을 그냥 함수의 인자로 넣어 재귀로 넣으면 함수내에서 따로 할당할 필요 없이 그냥 함수 call 할 때 바꿔서 쓰면 되며,
# chk_map 도, 함수의 처음과 끝에 적절히 배치 하면 함수 안에서 이것저것 신경 쓰지 않아도 된다.

T = int(input())

dr = [-1, 0, 1, 0]
dc = [0, 1, 0, -1]


def dfs(sr, sc, length, k_used):
    global max_length
    chk_map[sr][sc] = True

    max_length = max(max_length, length)

    for d in range(4):
        nr = sr + dr[d]
        nc = sc + dc[d]

        if nr < 0 or nc < 0 or nr >= n or nc >= n or chk_map[nr][nc]: continue
        if a[nr][nc] >= a[sr][sc]:
            if k_used: continue
            if a[nr][nc] - k < a[sr][sc]:
                or_value = a[nr][nc]
                a[nr][nc] = a[sr][sc] - 1
                dfs(nr, nc, length + 1, 1)
                a[nr][nc] = or_value

        elif a[nr][nc] < a[sr][sc]:
            dfs(nr, nc, length + 1, k_used)

    chk_map[sr][sc] = False

for t in range(T):
    n, k = map(int, input().split())
    a = [list(map(int, input().split())) for _ in range(n)]
    chk_map = [[False] * n for _ in range(n)]
    highest = 1
    coor_highest = []
    for r in range(n):
        for c in range(n):
            if a[r][c] > highest:
                coor_highest = [(r, c)]
                highest = a[r][c]
            elif a[r][c] == highest:
                coor_highest.append((r, c))

    max_length = 1
    cnt = 0
    for r, c in coor_highest:
        dfs(r, c, 1, 0)

    print('#{} {}'.format(t+1, max_length))
```

```python
# 백준 연구소

def spread_virus(walls):
    # 3개의 벽이 세워진 상태
    # 백트래킹은 지금까지 퍼진 바이러스의 양이 이미 구해진 최소 결과물보다 클 경우.
    # 결과물이 지금까지 중 제일 작다면 최소값 업데이트

    global min_spreaded_virus

    chk_map = [[False] * m for _ in range(n)]
    for i, j in walls:
        chk_map[i][j] = True

    q = []
    spreaded_virus_cnt = 0
    for i, j in viruses:
        chk_map[i][j] = True
        q.append((i, j))
    # 즉, 어차피 바이러스 순서가 전혀 중요하지 않기 때문에
    # 위처럼 큐에 바이러스를 모두 넣고 시작해도 된다.
    while q:
        sr, sc = q.pop()
        for d in range(4):
            nr = sr + dr[d]
            nc = sc + dc[d]

            if nr < 0 or nc < 0 or nr >= n or nc >= m: continue
            if chk_map[nr][nc] or a[nr][nc]: continue
            chk_map[nr][nc] = True
            spreaded_virus_cnt += 1
            q.append((nr, nc))

        if spreaded_virus_cnt > min_spreaded_virus:
            return

    min_spreaded_virus = min(min_spreaded_virus, spreaded_virus_cnt)

def build_three_walls(idx = 0, wall_cnt = 0, walls = []):
    if wall_cnt == 3:
        spread_virus(walls)
        return

    for zero in zeroes[idx:]:
        walls.append(zero)
        build_three_walls(idx+1, wall_cnt+1, walls)
        walls.pop()


n, m = map(int, input().split())

a = [list(map(int, input().split())) for _ in range(n)]

viruses = []
zeroes = []

for r in range(n):
    for c in range(m):
        if a[r][c] == 2:
            viruses.append((r, c))
        elif a[r][c] == 0:
            zeroes.append((r, c))

dr = [-1, 0, 1, 0]
dc = [0, 1, 0, -1]

min_spreaded_virus = 1e9

build_three_walls()

max_survived = len(zeroes) - 3 - min_spreaded_virus

print(max_survived)
```

```python
n, m = map(int, input().split())
sr, sc, sd = map(int, input().split())

a = [list(map(int, input().split())) for _ in range(n)]

dr = [-1, 0, 1, 0] # 북, 동, 남, 서. 회전방향은 왼쪽.
dc = [0, 1, 0, -1]

# 1. 현재위치
# 2-1. 왼쪽방향으로 돌면서 탐색 후, 해당 방향에 청소할 공간이 있으면
# 2-2. 그 방향으로 전진하고 청소
# 3. 네 번 방향을 바꿔도 청소할 곳이 없다면 방향을 유지한채로 후진
# 4. 후진한 상태에서 위 조건 반복
# 5. 청소할 곳도 없고 뒤쪽이 벽이라서 후진도 못하면 작동 중지

# 회전구현
def find_next_location(cr, cc, cd):
# 현재 방향 cd (index 순서와 일치)
    nd = cd
    for i in range(4):
        nd = (nd + 3) % 4 # 예를 들어 현재 1이라면, 0이 됨.
        nr = cr + dr[nd]
        nc = cc + dc[nd]
        if not a[nr][nc]:
            return nr, nc, nd, 0
    # 더 청소할 곳이 없는 경우 후진한다. 벽만 아니라면 후진 가능
    nd = (cd + 2) % 4 # 예로 1이라면 3 이 되고 3이라면 1이 됨
    nr = cr + dr[nd]
    nc = cc + dc[nd]
    if a[nr][nc] == 1: # 후진하려는 곳도 벽인 경우 -1 반환
        return -1, -1, -1, -1
    # 후진할 때는 기존 방향을 유지해야 한다.
    return nr, nc, cd, 1


a[sr][sc] = 2
area_cnt = 1 # 첫 칸은 무조건 청소할 수 있으므로 1로 시작

while True:
    sr, sc, sd, isback = find_next_location(sr, sc, sd)
    if sr == -1: break
    if isback:
        continue
    a[sr][sc] = 2
    area_cnt += 1

print(area_cnt)
```
